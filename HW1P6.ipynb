```python
import os
import pandas as pd
import seaborn
import matplotlib
%matplotlib inline

college = pd.read_csv('College.csv')

college.set_index('Names', inplace = True)

college.info()
college.describe()
college.head()
```

```python
seaborn.pairplot(college.iloc[:, 1:11])
```

```python
college.dropna(how='any', inplace=True)
college = college[college['Grad.Rate'] <= 100]

college.describe(include = 'all')
```

```python
seaborn.boxplot(college.iloc[:, 8], college.iloc[:, 0])
```

```python
bins = [0, 50, 100]
categories = ['No', 'Yes']

Elite = pd.cut(college['Top10perc'], bins, labels=categories)
college['Elite'] = pd.cut(college['Top10perc'], bins, labels=categories)
college['EliteBinned'] = pd.cut(college['Top10perc'], bins)

college.describe(include = 'all')
```

```python
seaborn.boxplot(college['Outstate'], college['Elite'])
```

```python
college['AcceptanceRate'] = college['Accept'] / college['Apps']
college.head()
```

```python
bins1 = [0, 5, 10, 15, 20]
categories1 = ['Very Low/Amazing', 'Low/Great', 'Medium/Okay', 'High/Poor']

SFRatioBinned = pd.cut(college['S.F.Ratio'], bins1, labels=categories1)
college['S.F.Ratio.Binned'] = pd.cut(college['S.F.Ratio'], bins1, labels=categories1)
college['S.F.Ratio.Bins'] = pd.cut(college['S.F.Ratio'], bins1)

college.describe(include = 'all')
```

```python
bins2 = [250, 1340.642214, 6800]
categories2 = ['Below Average', 'Above Average']

PersonalBinned = pd.cut(college['Personal'], bins2, labels=categories2)
college['PersonalBinned'] = pd.cut(college['Personal'], bins2, labels=categories2)
college['PersonalBins'] = pd.cut(college['Personal'], bins2)

college.describe(include = 'all')
```

```python
bins3 = [5000, 7500, 9660.171171, 20000, 60000]
categories3 = ['Very Low', 'Below Average', 'Good/Above Average', 'Outstanding']

ExpendBinned = pd.cut(college['Expend'], bins3, labels=categories3)
college['ExpendBinned'] = pd.cut(college['Expend'], bins3, labels=categories3)
college['ExpendBins'] = pd.cut(college['Expend'], bins3)

college.describe(include = 'all')
```

```python
seaborn.distplot(college['Expend'], [0, 5000, 7500, 9660.171171, 20000, 60000])
```

```python
seaborn.distplot(college['AcceptanceRate'])
```

```python
seaborn.distplot(college['Personal'], [250, 1340.642214, 2000, 6800])
```

```python
seaborn.barplot('Elite', 'AcceptanceRate', data = college)
# Elite colleges tend to accept less applicants than non-elite colleges.
```

```python
seaborn.barplot('Private', 'Apps', data = college)
# Students apply much more to public colleges than to private ones
```

```python
seaborn.lmplot(x = 'Apps', y = 'Accept', data = college)
# Logically, as more students apply, more are accepted.
```

```python
seaborn.barplot('Private', 'S.F.Ratio', data = college)
# Unsurprisingly, as public schools usually tend to be bigger than private ones, there are more
# students than faculty members.
```

```python
seaborn.lmplot(x = 'Accept', y = 'Enroll', data = college)
# It seems that most applicants who are accepted in a certain college enroll in that college.
```
